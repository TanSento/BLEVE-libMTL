```python
## Define the NN architecture
## NN with 3 hidden layer, s=[11, 256, 256] -> 8*[256,1]


class BleveNet(nn.Module):
    def __init__(self):
        super(BleveNet, self).__init__()
        # The first hidden layer has 256 neurons
        self.fc1 = nn.Linear(X_train_torch.shape[1], 256)
        # The second hidden layer has 256 neurons
        self.fc2 = nn.Linear(256, 256)
        
        # The final layer has 8 heads, each of which has 1 output neuron
        self.fc3 = nn.Linear(256, 256)
        self.fc4 = nn.Linear(256, 256)
        self.fc5 = nn.Linear(256, 256)
        self.fc6 = nn.Linear(256, 256)
        self.fc7 = nn.Linear(256, 256)
        self.fc8 = nn.Linear(256, 256)
        self.fc9 = nn.Linear(256, 256)
        self.fc10 = nn.Linear(256, 256)
        
        self.fc11 = nn.Linear(256,1)
        self.fc12 = nn.Linear(256,1)
        self.fc13 = nn.Linear(256,1)
        self.fc14 = nn.Linear(256,1)
        self.fc15 = nn.Linear(256,1)
        self.fc16 = nn.Linear(256,1)
        self.fc17 = nn.Linear(256,1)
        self.fc18 = nn.Linear(256,1)
       

        # Define proportion or neurons to dropout
        self.dropout = nn.Dropout(0.1)

        # Mish activation
        self.mish = nn.Mish()


    def forward(self, x):
        # add first fully connected layer
        x = self.fc1(x)
        x = self.mish(x)
        x = self.dropout(x)
        # add second fully connected layer
        x = self.fc2(x)
        x = self.mish(x)
        x = self.dropout(x)
       
        # separated in 8 different heads and outputs
        head1 = self.fc3(x)
        head1 = self.mish(head1)
        head1 = self.dropout(head1)
        output1 = self.fc11(head1)

        head2 = self.fc4(x)
        head2 = self.mish(head2)
        head2 = self.dropout(head2)
        output2 = self.fc12(head2)

        head3 = self.fc5(x)
        head3 = self.mish(head3)
        head3 = self.dropout(head3)
        output3 = self.fc13(head3)

        head4 = self.fc6(x)
        head4 = self.mish(head4)
        head4 = self.dropout(head4)
        output4 = self.fc14(head4)

        head5 = self.fc7(x)
        head5 = self.mish(head5)
        head5 = self.dropout(head5)
        output5 = self.fc15(head5)

        head6 = self.fc8(x)
        head6 = self.mish(head6)
        head6 = self.dropout(head6)
        output6 = self.fc16(head6)

        head7 = self.fc9(x)
        head7 = self.mish(head7)
        head7 = self.dropout(head7)
        output7 = self.fc17(head7)

        head8 = self.fc10(x)
        head8 = self.mish(head8)
        head8 = self.dropout(head8)
        output8 = self.fc18(head8)
        
       
        return [output1, output2, output3, output4, output5, output6, output7, output8]

# initialize the NN
model = BleveNet()
print(model)
```

```python
class UncertaintyWeightedLoss(nn.Module):
    def __init__(self, delta=1.0):
        super(UncertaintyWeightedLoss, self).__init__()

    def forward(self, losses, variances):
        weighted_loss = torch.mean(torch.div(losses, variances)) + torch.sum(torch.log(torch.sqrt(variances)))
        return weighted_loss
```

```python
loss_fn = UncertaintyWeightedLoss()

startTime = timeit.default_timer() 


n_epochs = 500

model.to(device)    # bring the model to gpu
quantiles = [quantile1, quantile2, quantile3, quantile4, quantile5, quantile6, quantile7, quantile8]
y_test = [y1_test, y2_test, y3_test, y4_test, y5_test, y6_test, y7_test, y8_test]

train_losses = []
lowest_mape = 100


for epoch in range(n_epochs):

    model.train()       
    # monitor training loss
    train_loss = 0.0
    
    ###################
    # train the model #
    ###################
    for data, target1, target2, target3, target4, target5, target6, target7, target8 in train_loader:
        # clear the gradients of all optimized variables
        optimizer.zero_grad()
        

      
        #bring data and target to gpu
        data, target1 = data.to(device), target1.to(device)
        target2, target3 = target2.to(device), target3.to(device)
        target4, target5 = target4.to(device), target5.to(device)
        target6, target7 = target6.to(device), target7.to(device)
        target8 = target8.to(device)


        with torch.no_grad():
            output1, output2, output3, output4, output5, output6, output7, output8 = model(data)
            var_1 = torch.var(output1).reshape(-1,1)
            var_2 = torch.var(output2).reshape(-1,1)
            var_3 = torch.var(output3).reshape(-1,1)
            var_4 = torch.var(output4).reshape(-1,1)
            var_5 = torch.var(output5).reshape(-1,1)
            var_6 = torch.var(output6).reshape(-1,1)
            var_7 = torch.var(output7).reshape(-1,1)
            var_8 = torch.var(output8).reshape(-1,1)
            variances = torch.cat((var_1, var_2, var_3, var_4, var_5, var_6, var_7, var_8), dim=1)

        
        # forward pass: compute predicted outputs by passing inputs to the model
        output1, output2, output3, output4, output5, output6, output7, output8 = model(data)



        # calculate the loss
        loss1 = criterion(output1, target1).reshape(-1,1)
        loss2 = criterion(output2, target2).reshape(-1,1)
        loss3 = criterion(output3, target3).reshape(-1,1)
        loss4 = criterion(output4, target4).reshape(-1,1)
        loss5 = criterion(output5, target5).reshape(-1,1)
        loss6 = criterion(output6, target6).reshape(-1,1)
        loss7 = criterion(output7, target7).reshape(-1,1)
        loss8 = criterion(output8, target8).reshape(-1,1)
        losses = torch.cat((loss1, loss2, loss3, loss4, loss5, loss6, loss7, loss8), dim=1)

        loss = loss_fn(losses, variances)
        
        # backward pass: compute gradient of the loss with respect to model parameters
        loss.backward()
        # perform a single optimization step (parameter update)
        optimizer.step()
        # update running training loss
        train_loss += loss.item()*data.size(0)
        
    # print training statistics 
    # calculate average loss over an epoch
    train_loss = train_loss/len(train_loader.dataset)
    train_losses.append(train_loss)

    print('Epoch: {}/{} \tTraining Loss: {:.6f}'.format(
        epoch+1,
        n_epochs, 
        train_loss
        ))
    
    
    ############################
    # Evaluate the lowest MAPE #
    ############################
    model.eval()
    with torch.no_grad():  # turn off gradient to save memory
        y_predNN_torch = model(X_test_torch.to(device))

    
    MAPE = 0
    for i in range(8):
        y_predNN = quantiles[i].inverse_transform(y_predNN_torch[i].cpu().numpy())
        mape = mean_absolute_percentage_error(y_test[i], y_predNN)
        MAPE += mape
    
    AVG_MAPE = MAPE/8
    if AVG_MAPE < lowest_mape:
        lowest_mape = AVG_MAPE
        print(f'Lowest MAPE: {lowest_mape}')
        torch.save(model.state_dict(), 'PyTorch_Model/UW-Multitask_lowest.pt')
    print()


    
endTime = timeit.default_timer()
runningTotal = (endTime - startTime)
print(f"Time taken: {runningTotal}")
```
