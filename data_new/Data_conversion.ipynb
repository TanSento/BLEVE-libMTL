# Automated Scripts

```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
writer_1 = pd.ExcelWriter('train.xlsx', engine='xlsxwriter')
writer_2 = pd.ExcelWriter('test.xlsx', engine='xlsxwriter')


target_name = ['positive_peak_time', 'negative_peak_time', 'arrival_time', 'positive_duration', 'negative_duration', 'positive_pressure', 'negative_pressure', 'positive_impulse']
for k in target_name:
    # Load the Excel file into a DataFrame
    df_read = pd.read_excel('BLEVE_Open_500Butan_500Propane_5-40m_8outputs.xlsx', sheet_name=k)
    df_full = df_read.iloc[:800,:]
    df_train, df_test = train_test_split(df_full, test_size=0.25, random_state=42)   # with random state, all sheets will be splitted in the same way


    df = df_train
    # Create a new DataFrame to store the split rows
    df_out = pd.DataFrame(columns=df.columns[:11])

    # Iterate over the rows in the original DataFrame
    for i in range(len(df)):
        for col in range(11, 47):
            # create a copy of the input DataFrame with only the first 11 columns and the current column, this takes all data-points
            df_col = df.iloc[:, list(range(11)) + [col]].copy()
            # rename the current column to "Target"
            df_col = df_col.rename(columns={df_col.columns[-1]: 'Target'})
            # repeat each row and append to the output DataFrame, we only refer to the current row i
            # df_out = df_out.append(df_col.iloc[[i]].reset_index(drop=True), ignore_index=True)            # append is deprecated
            df_out = pd.concat([df_out, df_col.iloc[[i]]], ignore_index=True)

    numbers = list(range(5,41))*600
    df_out["Stand-off Distance"] = numbers

    # swap the position of the 12th and 13th columns
    df_out.iloc[:, [11, 12]] = df_out.iloc[:, [12, 11]].to_numpy()    # convert column 12,11 into a 2-D numpy array to overwrite and overwrite them
    # Swap the column name
    df_out.columns = list(df_out.columns[:11]) + ['Stand-off Distance', 'Target']
    #Save
    
    df_out.to_excel(writer_1,index=False, sheet_name=k)
    


    #############################################################
    #############################################################

    df = df_test
    # Create a new DataFrame to store the split rows
    df_out = pd.DataFrame(columns=df.columns[:11])

    # Iterate over the rows in the original DataFrame
    for i in range(len(df)):
        for col in range(11, 47):
            # create a copy of the input DataFrame with only the first 11 columns and the current column, this takes all data-points
            df_col = df.iloc[:, list(range(11)) + [col]].copy()
            # rename the current column to "Target"
            df_col = df_col.rename(columns={df_col.columns[-1]: 'Target'})
            # repeat each row and append to the output DataFrame, we only refer to the current row i
            # df_out = df_out.append(df_col.iloc[[i]].reset_index(drop=True), ignore_index=True)            # append is deprecated
            df_out = pd.concat([df_out, df_col.iloc[[i]]], ignore_index=True)


    numbers = list(range(5,41))*200
    df_out["Stand-off Distance"] = numbers

    # swap the position of the 12th and 13th columns
    df_out.iloc[:, [11, 12]] = df_out.iloc[:, [12, 11]].to_numpy()    # convert column 12,11 into a 2-D numpy array to overwrite and overwrite them
    # Swap the column name
    df_out.columns = list(df_out.columns[:11]) + ['Stand-off Distance', 'Target']
    #Save
    df_out.to_excel(writer_2,index=False, sheet_name=k)

writer_1.save()
writer_2.save()
```
