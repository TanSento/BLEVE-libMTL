```python
import matplotlib.pyplot as plt
from numpy import save, load
import numpy as np
```

```python
weighting = 'RLW'
name = 'RLW-LR1e-3-RepgradFalse'
name_train_loss = f'RLW-LR1e-3-RepgradFalse_Train_Loss_Buffer'
name_test_loss = f'RLW-LR1e-3-RepgradFalse_Test_Loss_Buffer'
name_test_mape = f'RLW-LR1e-3-RepgradFalse_Test_MAPE_Buffer'


train_loss_buffer = load(f'buffer/{weighting}/{name_train_loss}.npy')
test_loss_buffer = load(f'buffer/{weighting}/{name_test_loss}.npy')
test_mape_buffer = load(f'buffer/{weighting}/{name_test_mape}.npy')
```

```python
fig, axs = plt.subplots(4, 2, figsize = (15,15))
            
axs[0,0].set_ylim([0,0.05])
axs[0,0].plot(train_loss_buffer[0], label = 'train loss')
axs[0,0].plot(test_loss_buffer[0], label = 'test loss')
axs[0,0].set_title("Positive Peaktime")
axs[0,0].set_xlabel('Epoch')
axs[0,0].set_ylabel('Loss Value')
axs[0,0].legend()


axs[0,1].set_ylim([0,0.05])
axs[0,1].plot(train_loss_buffer[1], label = 'train loss')
axs[0,1].plot(test_loss_buffer[1], label = 'test loss')
axs[0,1].set_title("Negative Peaktime")
axs[0,1].set_xlabel('Epoch')
axs[0,1].set_ylabel('Loss Value')
axs[0,1].legend()

axs[1,0].set_ylim([0,0.05])
axs[1,0].plot(train_loss_buffer[2], label = 'train loss')
axs[1,0].plot(test_loss_buffer[2], label = 'test loss')
axs[1,0].set_title("Arrival Time")
axs[1,0].set_xlabel('Epoch')
axs[1,0].set_ylabel('Loss Value')
axs[1,0].legend()

axs[1,1].set_ylim([0,0.05])
axs[1,1].plot(train_loss_buffer[3], label = 'train loss')
axs[1,1].plot(test_loss_buffer[3], label = 'test loss')
axs[1,1].set_title("Positive Duration")
axs[1,1].set_xlabel('Epoch')
axs[1,1].set_ylabel('Loss Value')
axs[1,1].legend()

axs[2,0].set_ylim([0,0.05])
axs[2,0].plot(train_loss_buffer[4], label = 'train loss')
axs[2,0].plot(test_loss_buffer[4], label = 'test loss')
axs[2,0].set_title("Negative Duration")
axs[2,0].set_xlabel('Epoch')
axs[2,0].set_ylabel('Loss Value')
axs[2,0].legend()

axs[2,1].set_ylim([0,0.05])
axs[2,1].plot(train_loss_buffer[5], label = 'train loss')
axs[2,1].plot(test_loss_buffer[5], label = 'test loss')
axs[2,1].set_title("Positive Pressure")
axs[2,1].set_xlabel('Epoch')
axs[2,1].set_ylabel('Loss Value')
axs[2,1].legend()

axs[3,0].set_ylim([0,0.05])
axs[3,0].plot(train_loss_buffer[6], label = 'train loss')
axs[3,0].plot(test_loss_buffer[6], label = 'test loss')
axs[3,0].set_title("Negative Pressure")
axs[3,0].set_xlabel('Epoch')
axs[3,0].set_ylabel('Loss Value')
axs[3,0].legend()

axs[3,1].set_ylim([0,0.05])
axs[3,1].plot(train_loss_buffer[7], label = 'train loss')
axs[3,1].plot(test_loss_buffer[7], label = 'test loss')
axs[3,1].set_title("Positive Impulse")
axs[3,1].set_xlabel('Epoch')
axs[3,1].set_ylabel('Loss Value')
axs[3,1].legend()


plt.suptitle(f"{name} Loss", fontsize="18")
plt.tight_layout()
# plt.savefig(f"Plots/{name}_Loss.png", dpi = 300)
plt.show()
# plt.clf()
```
