```python
# Create a new DataFrame to store the split rows
df_out = pd.DataFrame(columns=df.columns[:11])

# Iterate over the rows in the original DataFrame
for i in range(len(df)):
    for col in range(11, 47):
        # create a copy of the input DataFrame with only the first 11 columns and the current column, this takes all data-points
        df_col = df.iloc[:, list(range(11)) + [col]].copy()
        # rename the current column to "Target"
        df_col = df_col.rename(columns={df_col.columns[-1]: 'Target'})
        # repeat each row and append to the output DataFrame, we only refer to the current row i
        # df_out = df_out.append(df_col.iloc[[i]].reset_index(drop=True), ignore_index=True)            # append is deprecated
        df_out = pd.concat([df_out, df_col.iloc[[i]]], ignore_index=True)
```
